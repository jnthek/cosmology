```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as cst
```

```python
def occup_planck(nu,T):
    h = cst.Planck
    kB = cst.Boltzmann
    N = 1/(np.exp(h*nu/(kB*T)) - 1)
    return N
```

```python
def black_body_spec(nu, T):
    h = cst.Planck
    c = cst.speed_of_light
    I_nu = (2*h*nu**3/(c**2))*occup_planck(nu,T)
    return I_nu
```

```python
freq = np.linspace(1,300*1e9)

plt.figure()
plt.semilogy(freq, occup_planck(freq, 3))

plt.figure()
plt.plot(freq/1e9, black_body_spec(freq, 3))
```
