```python
import numpy as np
import scipy.constants as cst
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# wl = np.linspace(10,5000,5000)
# n1 = np.arange(1,10)
n2 = np.arange(2,100)
R = cst.Rydberg
wl_inv = R*((1/1)**2 - (1/n2)**2)
wl = 1/wl_inv
```

```python
# plt.figure(figsize=(16,9))
plt.stem(wl*1e10, np.ones(len(wl)), use_line_collection=True)
plt.plot(wl*1e10,wl*0.0)
plt.grid()
```
