```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as cst

ev2erg = 1.6e-12
```

```python
kB = cst.k_B.cgs.value
T = np.logspace(1,8,1000)
# T = 300
# n_n = 3e19
n_n_1 = 3
n_n_2 = 3e19
Ui = 14.5*ev2erg

n_i_1 = np.sqrt(n_n_1*2.4e15 * T**1.5 * np.exp(-Ui/(kB*T)))
n_i_2 = np.sqrt(n_n_2*2.4e15 * T**1.5 * np.exp(-Ui/(kB*T)))

# print (n_i)

plt.figure()
plt.loglog(T, n_i_1)
plt.loglog(T, n_i_2)
plt.ylim(1e-10, 1e50)
```
